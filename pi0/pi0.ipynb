```python
from root_pandas import read_root
import matplotlib.pyplot as plt
import glob
```

```python
listnames = ['pi0_eff20', 'pi0_eff30', 'pi0_eff40', 'pi0_eff50', 'pi0_eff60', 'pi0_all']
pi0 = {}
for name in listnames:
    pi0[name] = read_root('pi0.root', name)
```

```python
for name, df in pi0.items():
    plt.figure()
    plt.hist(df.M, bins = 200, range = [0.1, 0.17])
    plt.hist(df[df.isSignal == 0].M, bins = 200, range = [0.1, 0.17])
    plt.title(name)
```
